### Social Blade

```python
from selenium.webdriver.common.by import By # By es para buscar por tag, clase, id...

import time

import warnings
warnings.filterwarnings('ignore')
```

```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from selenium.webdriver.common.by import By
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC
from webdriver_manager.chrome import ChromeDriverManager
from selenium.webdriver.chrome.options import Options
```

```python
PATH = ChromeDriverManager().install()  # instala el driver de Chrome

# opciones del driver
opciones=Options()

# quita la bandera de ser robot
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)

# guardar las cookies
opciones.add_argument('cookies')    # mantiene las coockies

PATH =  ChromeDriverManager().install()

# URL de la página
url = "https://socialblade.com"

# Iniciar el navegador
driver=webdriver.Chrome(PATH, options=opciones)

# Cargar la página
driver.get(url)

# Esperar un tiempo para cargar la página
time.sleep(10)

#encontramos el elemento en el iframe(un html dentro de un html)
iframe=driver.find_element(By.XPATH, '//*[@id="sp_message_iframe_809946"]')
driver.switch_to.frame(iframe)

#hacemos click en el elemento
sub=driver.find_element(By.XPATH, '//*[@id="notice"]/div[5]/button[2]') # abre el desplegable
sub.click()

#volvemos al html principal
driver.switch_to.default_content()

#buscamos a los influencers

texto = driver.find_element(By.XPATH, '//*[@id="SearchInput"]')

texto.send_keys('gladoop ')   # escribo texto

buscar = driver.find_element(By.XPATH, '//*[@id="topSearchForm"]/div[1]/button')

buscar.click()
```

```python
tabla_1 = driver.find_element(By.XPATH,'//*[@id="YouTubeUserTopInfoBlock"]').text.split('\n')
tabla_1
```

```python
# Convertir la lista en un arreglo numpy y redimensionarlo a una matriz de 6x2
matriz = np.array(tabla_1).reshape((2, 6))

# Crear el DataFrame
df = pd.DataFrame(matriz, columns=['Atributo', 'Valor'])

# Mostrar el DataFrame
df
```

```python
tuplas_valores = [(tabla_1[i], tabla_1[i+1]) for i in range(0, len(tabla_1), 2)]
tuplas_valores
df = pd.DataFrame(tuplas_valores, columns = ["Atributo", "Valor"])
df
```
