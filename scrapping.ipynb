```python
from selenium import webdriver
from selenium.webdriver.chrome.options import Options
from selenium.webdriver.common.by import By
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import time

def obtener_informacion_influencer(nombre):
    try:
        # Código para obtener la información del influencer
        texto = driver.find_element(By.XPATH, '//*[@id="SearchInput"]')
        texto.clear()
        texto.send_keys(nombre)
        
        buscar = driver.find_element(By.XPATH, '//*[@id="topSearchForm"]/div[1]/button')
        buscar.click()
        
        tabla_2 = driver.find_element(By.XPATH, '//*[@id="YouTubeUserTopInfoBlock"]').text.split('\n')
        
        dictionary = {}
        for i in range(0, len(tabla_2), 2):
            key = tabla_2[i]
            value = tabla_2[i+1]
            dictionary[key] = value
        
        df_influencer = pd.DataFrame([dictionary])
        df_influencer['Gamer'] = nombre
        
        return df_influencer
    
    except Exception as e:
        print(f"Error al obtener información del influencer {nombre}: {str(e)}")
        return pd.DataFrame()


# Opciones del driver
opciones = Options()
opciones.add_experimental_option('excludeSwitches', ['enable-automation'])
opciones.add_experimental_option('useAutomationExtension', False)
opciones.add_argument('cookies')

# Path del driver de Chrome
PATH = ChromeDriverManager().install()

# URL de la página
url = "https://socialblade.com"

# Iniciar el navegador
driver = webdriver.Chrome(PATH, options=opciones)

# Cargar la página
driver.get(url)

# Esperar un tiempo para cargar la página
time.sleep(5)

# Encontramos el elemento en el iframe (un html dentro de otro html)
iframe = driver.find_element(By.XPATH, '//*[@id="sp_message_iframe_809946"]')
driver.switch_to.frame(iframe)

# Hacemos click en el elemento
sub = driver.find_element(By.XPATH, '//*[@id="notice"]/div[5]/button[2]') # abre el desplegable
sub.click()

# Volvemos al html principal
driver.switch_to.default_content()

# Lista de nombres de influencers
nombre = name_play_influencer

# DataFrame para almacenar los resultados
df = pd.DataFrame()

# Obtener información de cada influencer y añadir al DataFrame
for nombre in name_play_influencer:
    df_influencer = obtener_informacion_influencer(nombre)
    df = df.append(df_influencer, ignore_index=True)

# Cerrar el navegador
driver.quit()

# Imprimir el DataFrame con la información de los influencers
print(df)
```
